# 2.1 ip地址与mac地址
### 2.2.1 查看IP地址

```python
import subprocess
cmd = 'ipconfig'
p = subprocess.Popen(cmd, shell=True, stdout=subprocess.PIPE)
res = p.stdout.read().decode('gbk')
print(res)
```

#### 2.1.2查看mac地址

```python
import subprocess
cmd = 'ipconfig/all'
p = subprocess.Popen(cmd, shell=True, stdout=subprocess.PIPE)
res = p.stdout.read().decode('gbk')
print(res)
```

# 2.2 端口
#### 查看系统开放端口

```python
cmd = 'netstat -a -n'
p = subprocess.Popen(cmd, shell=True, stdout=subprocess.PIPE)
res = p.stdout.read().decode('gbk')
print(res)
```
